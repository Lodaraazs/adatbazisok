```sql
-- Az egyes egység megnevezések darabszáma ahol az egység nem "db"

SELECT  DISTINCT MEGYS AS 'Egység megnevezés',
        COUNT(*) AS 'DB/Egység megnevezés'
FROM Termek
WHERE MEGYS <> 'db'
GROUP BY MEGYS
```

```sql
-- A leghosszabb termékelnevezések

SELECT TOP 10 MEGNEVEZES AS 'Megnevezés', 
        LEN(MEGNEVEZES) AS 'Megnevezés hossza'
FROM Termek
GROUP BY MEGNEVEZES
ORDER BY 2 DESC
```

```sql
-- Az egyedi KAT_ID-k összege

SELECT COUNT(DISTINCT KAT_ID) AS 'Egyedi kategóriák összege'
FROM Termek
WHERE KAT_ID IS NOT NULL
```

```sql
-- A LISTAAR oszlop értékeinek statisztikája

SELECT MIN(LISTAAR) AS legalacsonyabb_egységár, 
       MAX(LISTAAR) AS legmagasabb_egységár, 
       AVG(LISTAAR) AS átlag_egységár, 
       STDEV(LISTAAR) AS egységár_szórása, 
       VAR(LISTAAR) AS egységár_szórásnégyzete
FROM Termek
WHERE LISTAAR IS NOT NULL
```

```sql
-- Hány darab NULL értéket tartalmazó sor található a LEIRAS nevű oszlopban?

SELECT  Sor_típusa,
        COUNT(*) Sor_darab
FROM
    (SELECT 
        CASE WHEN LEIRAS IS NULL THEN 'Null'
        ELSE 'Nem üres'
        END AS Sor_típusa
    FROM Termek) rows
GROUP BY Sor_típusa
```

```sql
-- A raktárban a legnagyobb összértékkel rendelkező termékek és összértékük

SELECT 
    TOP 5 MEGNEVEZES AS 'Termék neve',
    LISTAAR*KESZLET AS 'Összérték'
FROM Termek
GROUP BY MEGNEVEZES, LISTAAR*KESZLET
ORDER BY 2 DESC
```

```sql
-- Az első és utolsó adatfelvitel közt eltelt idő napokban

SELECT DATEDIFF(DAY, MIN(FELVITEL), MAX(FELVITEL))
FROM Termek
```
